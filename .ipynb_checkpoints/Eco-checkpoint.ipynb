### Feburary stats


```python
import pandas as pd
```

```python
raw_df_7th = pd.read_csv("EcoRing_February_7th.csv",index_col=0)
raw_df_14th = pd.read_csv("EcoRing_Feburuary_14th.csv",index_col=0)
raw_df_21th = pd.read_csv("EcoRing_Feburuary_21st.csv",index_col=0)
raw_df_21th
```

```python
raw_df_21th.loc[raw_df_7th["Brand Name"].isna()]
```

```python
df_7th = raw_df_7th.iloc[:30] 
df_14th = raw_df_14th.iloc[:15] 
df_21th = raw_df_21th.iloc[:20] 
feb_df = pd.concat([df_7th, df_14th, df_21th], ignore_index=True)
feb_df
```

###  Profit/Loss Analysis

```python
sold_items = feb_df[feb_df['Sold'] == True]
total_profit = sold_items['Profit / Loss'].sum()

unsold_items = feb_df[feb_df['Sold'] == False]
unsold_items["Profit / Loss"] =-440 
total_fee =  unsold_items["Profit / Loss"].sum()

print(total_profit)
feb_total = total_profit + total_fee
print(feb_total)
```

```python
sold_items['result(after tax)'] = sold_items['result(after tax)'].replace('[¥,]', '', regex=True).astype(float)
sold_items['Cost(price after Tax)'] = sold_items['Cost(price after Tax)'].replace('[¥,]', '', regex=True).astype(float)
```

```python
# Group by Brand Name and calculate total sales, total profit/loss, average price, and count of sold items
brand_sales = sold_items.groupby('Brand Name').agg(
    total_sales=('result(after tax)', 'sum'),
    total_profit_loss=('Profit / Loss', 'sum'),
    average_price=('Cost(price after Tax)', 'mean'),
    items_sold=('Brand Name', 'count')  # Counting the number of items sold per brand
)

# Calculate the total profit across all brands
total_profit = brand_sales['total_profit_loss'].sum()

# Calculate the percentage contribution of each brand's profit
brand_sales['profit_share_percentage'] = (brand_sales['total_profit_loss'] / total_profit) * 100

# Sort by total sales to see which brand has the highest sales
brand_sales_sorted = brand_sales.sort_values(by='total_sales', ascending=False)

brand_sales_sorted
```

```python
feb_df.to_csv("february_sales.csv", index=False)
```
